```python
from langchain_groq import ChatGroq
from ragas.llms import LangchainLLMWrapper
from langchain_ollama import OllamaEmbeddings,ChatOllama
import os
from ragas.testset import TestsetGenerator
from langchain_community.document_loaders import DirectoryLoader
from langchain_community.document_loaders import PyPDFLoader  # PDF-specific loader
```

```python
llm = ChatOllama(model="llama3.1:8b")
```

```python
generator_llm = LangchainLLMWrapper(llm)
embeddings = OllamaEmbeddings(model="nomic-embed-text")
```

```python
def load_pdfs(pdf_folder_path):
    documents = []
    for file in os.listdir(pdf_folder_path):
        if file.endswith('.pdf'):
            pdf_path = os.path.join(pdf_folder_path, file)
            loader = PyPDFLoader(pdf_path)
            documents.extend(loader.load())

    return documents
```

```python

# Path to the directory containing PDFs
path = "data/"

docs = load_pdfs(path)
```

```python
llm.invoke("What is the meaning of life?")
```

```python

generator = TestsetGenerator(llm=generator_llm, embedding_model=embeddings)
dataset = generator.generate_with_langchain_docs(docs, testset_size=10)
```
